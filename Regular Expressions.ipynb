```python
import re
```

```python
text = ""
with open("./T1.txt", 'r', encoding='UTF-8') as file:
    for line in file:
        text += line
abbrev = re.compile("[А-Я]{2,}")
match = re.findall(abbrev, text)
match
```

```python
abbrev = re.compile("\\[[^\]]*\\]")
match = re.findall(abbrev, text)
match
```

```python
abbrev = re.compile('«[^»]*\»')
match = re.findall(abbrev, text)
match
```

```python
abbrev = re.compile('Согласно [^,]*(?=,)')
match = re.findall(abbrev, text)
match
```

```python
abbrev = re.compile('[^.\n]* — это [^.]*\.')
match = re.findall(abbrev, text)
match
```

```python
abbrev = re.compile('(?<=[^ \n«…])[А-Я][а-я]+')
match = re.findall(abbrev, text)
match
```
