```python
#1: Extracting data
from os import listdir
filenames = ["./Clustering/" + n for n in listdir("./Clustering")]
texts = [[] for i in range(len(filenames))]
for i in range(len(filenames)):
    name = filenames[i]
    with open(name, 'r', encoding='UTF-8') as file:
        for line in file:
            texts[i].extend(line.split())
print(texts)
```

```python
#2: TF-IDF
import numpy as np
import math

words = list({w for t in texts for w in t})
dictionary = {}
for i in range(len(words)):
    dictionary[words[i]] = i
    
idf = np.zeros(len(dictionary))
for text in texts:
    for word in text:
        if word in dictionary:
            idf[dictionary[word]] += 1
            
idf = np.log(len(texts) / idf)
for i in range(len(dictionary)):
    if math.isinf(idf[i]):
        idf[i] = 0
        
def vectorize(text, dictionary):
    tf = np.zeros(len(dictionary))
    for w in text:
        if w in dictionary:
            tf[dictionary[w]] += 1
    tf /= len(text)
    return tf * idf

vectors = [vectorize(text, dictionary) for text in texts]

def max_n(vector, n):
    _max = [(0, 0)] * n
    for i in range(len(vector)):
        _max.append((vector[i], i))
        _max = sorted(_max, key=lambda t: t[0], reverse=True)
        _max = _max[:-1]
    return _max

max_10 = [max_n(v, 10) for v in vectors]
print(max_10)
```

```python
#3: extracting features
max_50 = [max_n(v, 50) for v in vectors]
features = set()
for i in range(len(max_50)):
    for t in max_50[i]:
        features.add(words[t[1]])
print(features)
```

```python
#4: Vectorizing texts
words = list({w for w in features})
_dict = {words[i]: i for i in range(len(words))}
def vectorize_text(text):
    vector = np.array([0] * len(words))
    for word in text:
        if word in _dict:
            vector[_dict[word]] = 1
    return vector
vectors = [vectorize_text(text) for text in texts]
print(vectors)
```

```python
#5: Similarity matrix
matrix = np.ndarray((len(texts), len(texts)))
for i in range(len(texts)):
    for j in range(len(texts)):
        matrix[i, j] = vectors[i] @ vectors[j]
print(matrix)
print(np.delete(matrix, 1, 1))
```

```python
#6: Clustering
cluster_names = [i for i in range(len(texts))]
while len(cluster_names) > 2:
    max_clusters = (0, 1)
    for i in range(len(cluster_names)):
        print(i)
        for j in range(len(cluster_names)):
            if i == j:
                continue
            new_max_clusters = (i, j)
            if matrix[i, j] > matrix[max_clusters[0], max_clusters[1]]:
                max_clusters = new_max_clusters
    i = max_clusters[0]
    j = max_clusters[1]
    
    vector = (matrix[i] + matrix[j]) / 2
    similarities = [0] * (len(cluster_names) - 1)
    
    vector = np.delete(vector, [i, j])
    matrix = np.delete(matrix, i, 0)
    matrix = np.delete(matrix, i, 1)
    matrix = np.delete(matrix, j, 0)
    matrix = np.delete(matrix, j, 1)
    for k in range(len(cluster_names) - 2):
        similarities[k] = matrix[k] @ vector
    similarities[-1] = vector @ vector
    matrix = np.append(matrix, np.array([0] * (len(similarities) - 1)).reshape(1, -1), 0)
    matrix = np.append(matrix, np.array([0] * (len(similarities))).reshape(-1, 1), 1)
    for k in range(len(similarities)):
        matrix[-1, k] = similarities[k]
        matrix[k, -1] = similarities[k]
    cluster_names.append([cluster_names[i], cluster_names[j]])
    name_i = cluster_names[i]
    name_j = cluster_names[j]
    cluster_names.remove(name_i)
    cluster_names.remove(name_j)
    
```
